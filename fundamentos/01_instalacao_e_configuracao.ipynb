#### Ambiente Virtual (venv)

Após criar seu projeto, é altamente recomendado criar um ambiente virtual. O ambiente virtual isola as dependências do seu projeto, evitando conflitos com outros projetos e protegendo o sistema de instalações globais desnecessárias.

Para criar um ambiente virtual, abra o terminal (atalho: `Ctrl + aspas`) e execute:

```bash
python -m venv venv
```

Isso criará uma pasta chamada `venv` no seu projeto, contendo o ambiente virtual.

**Ativando o ambiente virtual:**
- No Windows:
  ```bash
  .\venv\Scripts\activate
  ```
- No macOS/Linux:
  ```bash
  source venv/bin/activate
  ```

Quando o ambiente virtual estiver ativado, você verá o nome dele no início da linha do terminal. Agora, todas as bibliotecas instaladas com `pip` ficarão restritas ao seu projeto.

**Para desativar o ambiente virtual:**
```bash
deactivate
```

Utilizar ambientes virtuais é uma boa prática em Python e facilita a organização e portabilidade dos seus projetos.

> **Observação:**  
> No windows usamos o powershell como editor, e usamos os comandos para windows, no codespaces estamos usando o linux e rodamos comandos do linux

### Google Colab

Se optar por usar o Colab, basta acessar https://colab.research.google.com/. Crie um notebook e comece a explorar. A maior parte do uso é semelhante ao VS Code, mas a principal diferença está na instalação de pacotes:

- **Google Colab:**
    ```bash
    !pip install <nome_do_pacote>
    ```

- **VS Code (dentro de um notebook):**
    ```bash
    %pip install <nome_do_pacote>
    ```

- **Terminal (fora do VS Code ou Colab):**
    ```bash
    pip install <nome_do_pacote>
    ```

    Possivelmente no início, você ache mais fácil trabalhar no colab, mas como futuramente trabalharemos com arquivos, recomendo aprender a usar via vs code até o fim deste módulo
